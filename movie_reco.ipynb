## Importing Basic Libraries

```python
import numpy as np 
import pandas as pd 
```

### Importing Dataset from csv files

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

### Merging both dataframes

```python
df = movies.merge(credits,on='title')
```

```python
df.columns
```

```python
df1 = df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
df1.dropna(inplace=True)
```

```python
df1
```

### Creating a function to convert genres & keywords into a list.

```python
import ast
from ast import literal_eval
```

```python
def convert(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
df1['genres'] = df1['genres'].apply(convert)
```

```python
df1['keywords'] = df1['keywords'].apply(convert)
```

```python
df1.head()
```

### Function for extracting top 3 casts

```python
def convert_top3(obj):
    List = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            List.append(i['name'])
            counter += 1
        else:
            break
    return List
```

```python
df1['cast'] = df1['cast'].apply(convert_top3)
```

### Function for extracting director of the movie.

```python
def convert_dir(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            List.append(i['name'])
            break
    return List
```

```python
df1['crew'] = df1['crew'].apply(convert_dir)
```

## Cleaning the data

```python
df1['overview'] = df1['overview'].apply(lambda x:x.split())
```

```python
df1['genres']= df1['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df1['keywords']= df1['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
df1['cast']= df1['cast'].apply(lambda x:[i.replace(" ","")for i in x])
df1['crew']= df1['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

## Merging all feature columns into one.

```python
df1['tags'] = df1['overview'] + df1['genres'] + df1['keywords'] + df1['cast'] + df1['crew']
```

```python
new_df = df1[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

## Applying stemming to avoid repeatable words in the feature column.

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## Converting tags column into vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

## Calculating distances between each vectors to know similarities between movies

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

## Main function

### To recommend movies we are sorting the vectors in descending order according to the given Movie titles's vector, and only top 5 vector's movie title.

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Predator')
```
